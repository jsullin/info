Install WSL (if you haven't already):

Open PowerShell as Administrator
Run: wsl --install
Restart your computer
Follow the setup prompts to create a Linux username and password


Open WSL:

Type wsl in your Windows Command Prompt, or
Search for "Ubuntu" (or whatever Linux distro was installed) in the Start menu, or
Type bash in Command Prompt

RUN IN WSL: apk update && apk add curl bash


RUN: BASH

CURL commands will now work!
